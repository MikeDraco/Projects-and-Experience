```
from pyspark.ml.feature import RegexTokenizer
rt = RegexTokenizer().setInputCol("text").setOutputCol("text_out").setPattern(" ").setToLowercase(True)

tokenized=rt.transform(YelpDF_Binary.select(Columns_to_keep))
```

```
from pyspark.ml.feature import StopWordsRemover
englishStopWords = StopWordsRemover.loadDefaultStopWords("english")
stops = StopWordsRemover().setStopWords(englishStopWords).setInputCol('text_out')

text=stops.transform(tokenized)
```

```
tfIdfIn = tokenized
```

```
from pyspark.ml.feature import HashingTF, IDF
tf = HashingTF().setInputCol('text_out').setOutputCol("TFOut").setNumFeatures(100000)

idf = IDF().setInputCol("TFOut").setOutputCol("IDFOut").setMinDocFreq(2)

YelpDF_Binary=idf.fit(tf.transform(tfIdfIn)).transform(tf.transform(tfIdfIn))
```

```
display(YelpDF_Binary)
```

```
import pyspark
import pandas as pd
from pyspark.sql.functions import col
from pyspark.ml import Pipeline
from pyspark.sql.types import IntegerType
from pyspark.ml.feature import StringIndexer, VectorAssembler, OneHotEncoderEstimator

categoricalColumns = [field for (field, dataType) in YelpDF_Binary.dtypes 
                     if dataType == "string"]

categoricalColumns.remove('text')

stages = [] # stages in our Pipeline
for categoricalCol in categoricalColumns:
    # Category Indexing with StringIndexer
     stringIndexer = StringIndexer(inputCol=categoricalCol, outputCol=categoricalCol + "Index")
    # Use OneHotEncoder to convert categorical variables into binary SparseVectors
     encoder = OneHotEncoderEstimator(inputCols=[stringIndexer.getOutputCol()], outputCols=[categoricalCol + "OHE"])
    # Add stages.  These are not run here, but will run all at once later on.
     stages += [stringIndexer, encoder]
      
numericCols = [field for (field, dataType) in YelpDF_Binary.dtypes if (dataType == "double")]
  
numericCols.remove('label')
      
assemblerInputs = [c + "OHE" for c in categoricalColumns] + numericCols+['IDFOut']
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
stages += [assembler]

partialPipeline = Pipeline().setStages(stages)
pipelineModel = partialPipeline.fit(YelpDF_Binary)
FinalDF = pipelineModel.transform(YelpDF_Binary)
```

#Split the Data into Training and Testing

```
(trainingData, testData) = FinalDF.randomSplit([0.7, 0.3], seed=100)
print(trainingData.count())
print(testData.count())
```

#Logistic Regression

```
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Create initial LogisticRegression model
lr = LogisticRegression(labelCol="label", featuresCol="features", maxIter=10)

# Train model with Training Data
lrModel = lr.fit(trainingData)

# Make predictions on test data using the transform() method.

lrPrediction = lrModel.transform(testData)

# Evaluate model using AreaUnderROC
evaluator_roc = BinaryClassificationEvaluator()
areaUnderROC=evaluator_roc.evaluate(lrPrediction)
print("areaUnderROC is ", areaUnderROC)

# Evaluate model using AreaUnderPR
evaluator_pr=BinaryClassificationEvaluator(metricName="areaUnderPR")
areaUnderPR=evaluator_pr.evaluate(lrPrediction)
print("areaUnderPR is ", areaUnderPR)
```

###Evaluate the performance of the model

```
def print_performance(df):
  
  print("prediction accuracy is: ", df.where("prediction==label").count()/df.count())

  tp=df.where("label=1 and prediction=1").count()
  fp=df.where("label=0 and prediction=1").count()
  tn=df.where("label=0 and prediction=0").count()
  fn=df.where("label=1 and prediction=0").count()

  precision=tp/(tp+fp)
  
  recall=tp/(tp+fn)
  
  f=(2*precision * recall)/(precision + recall)

  print("true positive is: ", tp)

  print("false positive is: ", fp)

  print("true negative is: ", tn)

  print("false negative is: ", fn)

  print("precision is: ", precision) 

  print("recall is: ", recall)

  print("f1 score is: ", f)
  
```

#Lift Chart

```
def lift (test, pred, cardinaility):

    res = pd.DataFrame(np.column_stack((test, pred)),
                       columns=['Target','PR_0', 'PR_1'])

    res['scr_grp'] = pd.qcut(res['PR_0'], cardinaility, labels=False)+1

    crt = pd.crosstab(res.scr_grp, res.Target).reset_index()
    crt = crt.rename(columns= {'Target':'Np',0.0: 'Negatives', 1.0: 'Positives'})

    G = crt['Positives'].sum()
    B = crt['Negatives'].sum()
   
    avg_resp_rate = G/(G+B)

    crt['resp_rate'] = crt['Positives']/(crt['Positives']+crt['Negatives'])
    crt['lift'] = (crt['resp_rate']/avg_resp_rate)
    crt['rand_resp'] = 1/cardinaility
    crt['cmltv_p'] = (crt['Positives']).cumsum()
    crt['cmltv_p_perc'] = (crt['Positives']/G).cumsum()*100
    crt['cmltv_n'] = (crt['Negatives']).cumsum()  
    crt['cmltv_n_perc'] = (crt['Negatives']/B).cumsum()*100  
    crt['cmltv_rand_p_perc'] = (crt.rand_resp.cumsum())*100
    crt['cmltv_resp_rate'] = crt['cmltv_p']/(crt['cmltv_p']+crt['cmltv_n']) 
    crt['cmltv_lift'] = crt['cmltv_resp_rate']/avg_resp_rate
    crt['KS']=crt['cmltv_p_perc']-crt['cmltv_rand_p_perc']
    crt = crt.drop(['rand_resp','cmltv_p','cmltv_n',], axis=1)
    
    print('average response rate: ' , avg_resp_rate)
    return crt
```

```
import numpy as np
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType

firstelement=udf(lambda v:float(v[0]),FloatType())
prob_0=lrPrediction.select(firstelement('probability'))

secondelement=udf(lambda v:float(v[1]),FloatType())
prob_1=lrPrediction.select(secondelement('probability'))

prob_1=prob_1.toPandas()
prob_0=prob_0.toPandas()

prob=pd.concat([prob_0,prob_1],axis=1)
prob=prob.to_numpy()

Y_test= np.array(testData.select('label').collect())
```

#Cross Validation

```
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

# Create ParamGrid for Cross Validation
lr_paramGrid = (ParamGridBuilder()
             .addGrid(lr.regParam, [0.01, 1, 2.0])
             .addGrid(lr.elasticNetParam, [0.01,0.1, 0.5, 1.0])
             .build())

# Create 5-fold CrossValidator
cv_lr = CrossValidator(estimator=lr, estimatorParamMaps=lr_paramGrid, evaluator=evaluator_roc, numFolds=3)
```

```
# Run cross validations
# this will likely take a fair amount of time because of the amount of models we are creating and testing 

cv_lrModel = cv_lr.fit(trainingData)

# make prediction based on testData. cv_lrModel uses the best model found from the Cross Validation
cv_lrPrediction = cv_lrModel.transform(testData)

#Evaluate best model using areaUnderROC
evaluator_roc.evaluate(cv_lrPrediction)
```
