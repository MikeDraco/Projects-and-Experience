# Spotify Tweets from January 1st, 2020

```python
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd

begin_date=dt.date(2020,1,1)
end_date=dt.date(2020,4,30)

limit=7500
lang= 'english'

tweets=query_tweets('Spotify',begindate=begin_date,enddate=end_date,limit=limit,lang=lang)
SpotDf=pd.DataFrame(t.__dict__ for t in tweets )
```

# Wordcloud For Apple Music and Spotify Tweets

In this section we generated a wordcloud on all the tweets we collected. One thing we found is that most tweets that mentioned Apple Music also mentioned Spotify and vice-versa

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS

AppleSpotTweetsText=list(AppleDf['text'])+list(SpotDf['text'])
AppleSpotTweetsText=[x.split(' ') for x in AppleSpotTweetsText]
AppleSpotTweetsText=[x for x in AppleSpotTweetsText]
output=[]
def reemovNestings(x): 
    for i in x: 
        if type(i) == list: 
            reemovNestings(i) 
        else: 
            output.append(i) 
reemovNestings(AppleTweetsText)

try:
    while True:
        output.remove('')
except ValueError:
    pass

word_counts = {} # Initialize a Dictionary
for text in output:  #or the cleaned up text
    text = str(text).lower() # Convert to string
    for word in text.split(): # text.split() 
        word_counts[word] = word_counts.get(word, 0) + 1

stopwords = set(STOPWORDS)
delete_present_stopwords=[key for key in word_counts if key in stopwords]
for key in delete_present_stopwords:
    del word_counts[key]
    
wordcloud = WordCloud(stopwords=stopwords, width = 1500, height = 1000, 
                background_color ='white', 
                min_font_size = 10).generate_from_frequencies(word_counts)

plt.figure(figsize = (14, 14), facecolor = None) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.tight_layout(pad = 0)
plt.show()
```

## Apple Sentiment Scores and Labels

```python
from textblob import TextBlob

AppleDf['text'] = AppleDf['text'].astype('str')

AppleDf['PolarityScore'] = AppleDf['text'].apply(lambda txt:  TextBlob(txt).polarity)
AppleDf['SubjectivityScore'] = AppleDf['text'].apply(lambda txt:  TextBlob(txt).subjectivity)

def sentiment_bins(data):
    if data <= -0.5:
        grouping = 'Strong-Negative'
    elif data > -0.5 and data < 0.0:
        grouping = 'Mild-Negative'  
    elif data > 0.0 and data < 0.5:
        grouping = 'Mild Positive'
    elif data >=0.5:
        grouping = 'Strong-Positive'
    else:
        grouping = 'Neutral'
    return grouping

AppleDf['sentimentLabel'] = AppleDf['PolarityScore'].apply(sentiment_bins)
```

```python
AppleDf['sentimentLabel'].value_counts()
```

## Spotify Sentiment Scores and Labels

```python
SpotDf['text'] = SpotDf['text'].astype('str')

SpotDf['PolarityScore'] = SpotDf['text'].apply(lambda txt:  TextBlob(txt).polarity)
SpotDf['SubjectivityScore'] = SpotDf['text'].apply(lambda txt:  TextBlob(txt).subjectivity)

def sentiment_bins(data):
    if data <= -0.5:
        grouping = 'Strong-Negative'
    elif data > -0.5 and data < 0.0:
        grouping = 'Mild-Negative'  
    elif data > 0.0 and data < 0.5:
        grouping = 'Mild Positive'
    elif data >=0.5:
        grouping = 'Strong-Positive'
    else:
        grouping = 'Neutral'
    return grouping

SpotDf['sentimentLabel'] = SpotDf['PolarityScore'].apply(sentiment_bins)
```

```python
SpotDf['sentimentLabel'].value_counts()
```

Lastly, we plotted the polarity score against each stock. There is barely any correlation except 3/24 for spotify where we see a jump in both polarity and stock.

```python
AAPL=df[df['Apple/Spotify']=='Apple'].groupby(['Date']).mean()
AAPL.reset_index(inplace=True)
AppleData['Date']=AppleData['Date'].astype('str')
AppleStockAndTweets= pd.merge(AppleData,AAPL, on='Date')

import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.plot(AppleStockAndTweets['Date'],AppleStockAndTweets['Close'])
ax.set_xticklabels(AppleStockAndTweets['Date'],rotation=45)
ax.set_ylabel('Stock Market',color='blue')
ax.set_xlabel('Time')
ax2=ax.twinx()
ax2.plot(AppleStockAndTweets['PolarityScore'], color='orange')
ax2.set_ylim(-.5,.5)
ax2.set_ylabel('Polarity',color='orange')
ax.set_title('Apple Stock vs. Sentiment Score', fontsize=20)
```
